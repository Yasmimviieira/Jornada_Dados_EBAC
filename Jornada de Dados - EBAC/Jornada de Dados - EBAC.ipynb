# Processamento

```python
data = list()

with open(file="/content/drive/MyDrive/IT/Dados/Jornada de Dados - EBAC/State_of_data_2022.csv", mode='r') as db_principal:
  header = db_principal.readline()
  line = db_principal.readline()

  while line:
    line = line.split(sep=',')
    data.append({'Cargo': line[21], 'Experiencia': line[22], 'Salario': line[23]})
    line = db_principal.readline()
```

```python
for datum in data[0:10]:
  display(datum)
```

```python
len(data)
```

```python
import pandas as pd
```

```python
data_table = pd.DataFrame(data)
```

```python
data_table.head(n=15)
```

```python
data_table.shape
```

```python
data_table.to_csv("data.csv", header=True, index=False)
```

# Limpeza

* Linhas vazias

Existem diversas estratégias para lidar com dados faltantes, como a inputação ("forçar
valor") ou a deleção (remover a linha toda).

```python
data_table = data_table.replace('', pd.NA) # Substituir o que for igual a nada para NA
```

```python
data_table.head(n=15)
```

```python
data_table = data_table.dropna()
```

```python
data_table.head(n=15)
```

```python
data_table.shape
```

*  Linhas corrompidas

Linhas corrompidas podem derivar de diversos fatores: erros de processamento, erros
humanos, etc.

```python
data_table['Salario'] = data_table['Salario'].apply(lambda row: row if 'R$' in row else None)
```

```python
data_table.head(n=15)
```

```python
data_table = data_table.dropna()
```

```python
data_table.head(n=15)
```

```python
data_table.shape
```

*  Linhas de interesse

A seleção dos dados de interesse também pode ocorrer a nível de linha. Vamos selecionar apenas as profissões de interesse: **analista**, **cientista** e **engenheira de dados**.



```python
data_table['Cargo'].unique()
```

```python
positions = ['Analista de Dados/Data Analyst', 'Cientista de Dados/Data Scientist', 'Engenheiro de Dados/Data Engineer']
```

```python
data_table.loc[:, 'Cargo'] = data_table['Cargo'].apply(lambda row: row if row in positions else None)
```

```python
data_table.head(n=15)
```

```python
data_table = data_table.dropna()
```

```python
data_table.head(n=15)
```

```python
data_table.shape
```

Para simplificar um pouco mais, podemos renomear as profissões.

```python
labels = ['Analista', 'Cientista', 'Engenheira']
labels = dict(zip(positions, labels))
```

```python
display(labels)
```

```python
data_table['Cargo'] = data_table['Cargo'].apply(lambda row: labels[row])
```

# Por salário

```python
salario_agg = data_table[['Salario','Quantidade']].groupby('Salario').agg('sum').reset_index()
```

```python
salario_agg.head(n=15)
```

Ordenar os salários

```python
salario_ordem = {
'Menos de R$ 1.000/mês': 0,
'de R$ 1.001/mês a R$ 2.000/mês': 1,
'de R$ 2.001/mês a R$ 3.000/mês': 2,
'de R$ 3.001/mês a R$ 4.000/mês': 3,
'de R$ 4.001/mês a R$ 6.000/mês': 4,
'de R$ 6.001/mês a R$ 8.000/mês': 5,
'de R$ 8.001/mês a R$ 12.000/mês': 6,
'de R$ 12.001/mês a R$ 16.000/mês': 7,
'de R$ 16.001/mês a R$ 20.000/mês': 8,
'de R$ 20.001/mês a R$ 25.000/mês': 9,
'de R$ 25.001/mês a R$ 30.000/mês': 10,
'de R$ 30.001/mês a R$ 40.000/mês': 11,
'Acima de R$ 40.001/mês': 12
}
```

```python
salario_agg['Ordem'] = salario_agg['Salario'].apply(lambda row: salario_ordem[row])
```

```python
salario_agg = salario_agg.sort_values(by='Ordem', ascending=True)
```

```python
salario_agg.head(n=15)
```

# Por salário e cargo

No segundo nível de agregação, vamos contar quantos profissionais estão em cada faixa de salário por cargo.

```python
salario_cargo_agg = data_table[['Cargo', 'Salario', 'Quantidade']].groupby(['Cargo', 'Salario']).agg('sum').reset_index()
```

```python
salario_cargo_agg['Ordem'] = salario_cargo_agg['Salario'].apply(lambda row: salario_ordem[row])
```

```python
salario_cargo_agg = salario_cargo_agg.sort_values(by=['Cargo', 'Ordem'], ascending=[True, True])
```

```python
salario_cargo_agg.head(100)
```

Analisar o salário dos Analistas de Dados.

```python
salario_cargo_agg[salario_cargo_agg['Cargo'] == 'Analista'].head(n=15)
```

# Por salário, cargo e experiência

Profissionais estão em cada faixa de salário por cargo e nivel de experiência.

```python
salario_cargo_experiencia_agg = data_table[['Cargo', 'Experiencia', 'Salario', 'Quantidade']].groupby(['Cargo', 'Experiencia', 'Salario']).agg('sum').reset_index()
```

```python
salario_cargo_experiencia_agg['Ordem'] = salario_cargo_experiencia_agg['Salario'].apply(lambda row: salario_ordem[row])
```

```python
salario_cargo_experiencia_agg = salario_cargo_experiencia_agg.sort_values(by=['Cargo', 'Experiencia', 'Ordem'], ascending=[True, True, True])
```

```python
salario_cargo_experiencia_agg.head(n=15)
```

```python
salario_cargo_experiencia_agg[salario_cargo_experiencia_agg['Cargo'] == 'Analista'].head(n=30)
```
